```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

## Categorical Data

### a. countplot
**Use any one of the below**

```python
sns.countplot(x=df['Survived'], dodge=False)
# df['Survived'].value_counts().plot(kind = 'bar')
```

```python
sns.countplot(x=df['Pclass'])
```

```python
sns.countplot(x=df['Sex'])
```

### b. Piechart

```python
df['Survived'].value_counts().plot(kind = 'pie', autopct='%.2f')
```

## Numerical Data

### Histogram
**It displays the frequency or count of data points falling within specified intervals or bins.**

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

### Distplot

```python
# sns.displot(df['Age'])
sns.histplot(df['Age'], kde=True)
```

### Boxplot

```python
sns.boxplot(df['Age'])
```
